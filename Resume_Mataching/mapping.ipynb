```python
def read_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages:
            text += page.extract_text()
    return text
```

```python
# Read the resume PDF and preprocess
job_description = "Looking for a software engineer with 3+ years of experience, 2-4 yrs experience, and 5 years experience."

# Read job description
job_text = job_description.lower()
resume_path = 'resume (4).pdf'
resume_text = read_pdf(resume_path).lower()

# Extract experience from job description using regex
job_experience_matches = re.findall(r'(\d+)[ -]*((?:\+|\-)?\d*)\s*yr[s]*', job_text)

# Convert various experience formats to a single value
job_experience_years = []
for match in job_experience_matches:
    years = int(match[0])
    if match[1]:
        modifier = 1 if match[1] == '+' else -1
        modifier_years = int(match[1])
        years += modifier * modifier_years
    job_experience_years.append(years)

# Extract work experience section from resume using regex
work_experience_section = re.findall(r'work experience:(.*?)(?:\n\n|\n|$)', resume_text, re.DOTALL)
work_experience_text = ' '.join(work_experience_section).strip()

# Extract years from work experience section using regex
experience_years_matches = re.findall(r'\b\d{4}\b', work_experience_text)
experience_years = list(map(int, experience_years_matches))

# Classify experience years as start and end years
start_years = experience_years[::2]
end_years = experience_years[1::2]

# Calculate total experience based on start and end years
total_experience = sum(end_year - start_year + 1 for start_year, end_year in zip(start_years, end_years))

# Calculate experience score
max_job_experience = max(job_experience_years)
experience_score = min(total_experience / max_job_experience, 1) if max_job_experience > 0 else 0

print(f"Experience Score: {experience_score:.2f}")
```

```python
import re
import pandas as pd

def extract_data(resume_text):
    """
    Extract structured data from the resume text.
    
    Args:
        resume_text (str): The raw text extracted from the PDF.
        
    Returns:
        dict: A dictionary containing structured data.
    """
    data = {}

    # Extract Name
    name_match = re.search(r"^\s*(\w+\s+\w+)", resume_text)
    data['Name'] = name_match.group(1) if name_match else None

    # Extract Contact Information
    phone_match = re.search(r"\+?\d{10,13}", resume_text)
    data['Phone'] = phone_match.group(0) if phone_match else None
    
    email_match = re.search(r"[\w\.-]+@[\w\.-]+", resume_text)
    data['Email'] = email_match.group(0) if email_match else None
    
    linkedin_match = re.search(r"https?://[\w\.\/]+", resume_text)
    data['LinkedIn'] = linkedin_match.group(0) if linkedin_match else None

    # Extract Summary
    summary_match = re.search(r"Summary\s+(.*?)(?=Education|Work experience|Qualifications)", resume_text, re.DOTALL)
    data['Summary'] = summary_match.group(1).strip() if summary_match else None

    # Extract Education
    education_match = re.search(r"Education\s+(.*?)(?=Work experience|Qualifications)", resume_text, re.DOTALL)
    data['Education'] = education_match.group(1).strip() if education_match else None

    # Extract Work Experience
    experience_match = re.search(r"Work experience\s+(.*?)(?=Qualifications|Projects)", resume_text, re.DOTALL)
    data['Experience'] = experience_match.group(1).strip() if experience_match else None

    # Extract Qualifications
    qualifications_match = re.search(r"Qualifications\s+(.*?)(?=Projects|Skills)", resume_text, re.DOTALL)
    data['Qualifications'] = qualifications_match.group(1).strip() if qualifications_match else None

    # Extract Projects
    projects_match = re.search(r"Projects\s+(.*?)(?=Skills|References)", resume_text, re.DOTALL)
    data['Projects'] = projects_match.group(1).strip() if projects_match else None

    # Extract Skills
    skills_match = re.search(r"Skills\s+(.*?)(?=References)", resume_text, re.DOTALL)
    data['Skills'] = skills_match.group(1).strip() if skills_match else None

    return data


def save_to_csv(data_list, output_file):
    """
    Save extracted data to a CSV file.
    
    Args:
        data_list (list): A list of dictionaries containing structured data.
        output_file (str): The path to the output CSV file.
    """
    df = pd.DataFrame(data_list)
    df.to_csv(output_file, index=False)
    print(f"Data successfully saved to {output_file}")


# Sample extracted text (replace with actual extracted text)
extracted_text = """
Hassan Tahir
Machine Learning Engineer

https://11a55an.github.io

+923034624939

Model Town, Sambrial iamhassantahir@gmail.com

+923034624939

Summary

Results-driven Data Scientist and Senior Data Analyst with a strong background in machine learning and data analysis. Proven expertise in developing predictive models, conducting data-driven analyses, and collaborating with cross-functional teams to drive business growth. Passionate about solving complex problems using data-driven insights.

Education

BS Computer Science

2018 - 2022

University of Gujrat

FSc Pre Engineering

Punjab College Sambrial

2016 - 2018

Matriculation

Allied School Husnain Campus

2014 - 2016

Work experience

Navigation and Communication Solutions Pvt Ltd

2023 - Present

Artificial Intelligence Engineer

S&P Global

January 2022 - December 2022

Data Analyst

Qualifications

Certified IBM Data Scientist
Certified IBM Machine Learning Engineer
Certified IBM AI Engineering Specialist
Projects

Skills

References

Flood Forecasting System
Deep Learning

August 2021 - January 2022

Sign Language Detection
Computer Vision

January 2020 - August 2021

Transformer-based Stock Price Prediction
Regression Analysis

August 2020 - January 2021

Music Recommendation System
Recommendation System

January 2020 - August 2020

Python Programming
Machine Learning Algorithms (Random Forests, K-means clustering)
Data Visualization (Matplotlib, Seaborn) Data
Preprocessing
Data Cleaning
Tensorflow
PyTorch
Computer Vision
Git
Problem Solving
Team Collaboration
Communication Skills
Mr. Najeeb Ur Rehman Assistant Professor at University of Gujrat Email: najeeb.rehman@uog.edu.pk Phone: +92333123456
"""

# Extract structured data
structured_data = extract_data(extracted_text)

# Save the structured data to a CSV file
save_to_csv([structured_data], "resume_data.csv")
```

```python
def extract_data1(resume_text):
    """
    Extract structured data from the resume text.
    
    Args:
        resume_text (str): The raw text extracted from the PDF.
        
    Returns:
        dict: A dictionary containing structured data.
    """
    data = {}

    # Extract Name
    name_match = re.search(r"^\s*(\w+\s+\w+)", resume_text)
    data['Name'] = name_match.group(1) if name_match else None

    # Extract Contact Information
    phone_match = re.search(r"\+?\d{10,13}", resume_text)
    data['Phone'] = phone_match.group(0) if phone_match else None
    
    email_match = re.search(r"[\w\.-]+@[\w\.-]+", resume_text)
    data['Email'] = email_match.group(0) if email_match else None
    
    linkedin_match = re.search(r"https?://[\w\.\/]+", resume_text)
    data['LinkedIn'] = linkedin_match.group(0) if linkedin_match else None

    # Extract Summary
    summary_match = re.search(r"Summary\s+(.*?)(?=Education|Work experience|Qualifications)", resume_text, re.DOTALL)
    data['Summary'] = summary_match.group(1).strip() if summary_match else None

    # Extract Education
    education_match = re.search(r"Education\s+(.*?)(?=Work experience|Qualifications)", resume_text, re.DOTALL)
    data['Education'] = education_match.group(1).strip() if education_match else None

    # Extract Work Experience
    experience_match = re.search(r"Work experience\s+(.*?)(?=Qualifications|Projects)", resume_text, re.DOTALL)
    data['Experience'] = experience_match.group(1).strip() if experience_match else None

    # Extract Qualifications
    qualifications_match = re.search(r"Qualifications\s+(.*?)(?=Projects|Skills)", resume_text, re.DOTALL)
    data['Qualifications'] = qualifications_match.group(1).strip() if qualifications_match else None

    # Extract Projects
    projects_match = re.search(r"Projects\s+(.*?)(?=Skills|References)", resume_text, re.DOTALL)
    if projects_match:
        projects_raw = projects_match.group(1).strip()
        projects = re.findall(r"(.*?)\s+(.*?)\s+\((.*?)\)", projects_raw, re.DOTALL)
        data['Projects'] = [f"{title} - {desc} ({duration})" for title, desc, duration in projects]
    else:
        data['Projects'] = None

    # Extract Skills
    skills_match = re.search(r"Skills\s+(.*?)(?=References)", resume_text, re.DOTALL)
    if skills_match:
        skills_raw = skills_match.group(1).strip()
        data['Skills'] = [skill.strip() for skill in re.split(r",|;|\n", skills_raw) if skill.strip()]
    else:
        data['Skills'] = None

    return data
```
